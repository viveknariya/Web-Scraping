```python
from selenium import webdriver
import os
import warnings

os.environ['PATH'] += 'C:\Driver\chromedriver.exe'

warnings.filterwarnings("ignore",category=DeprecationWarning)

driver = webdriver.Chrome(executable_path='C:\Driver\chromedriver.exe')
```

```python
driver.get('https://www.linkedin.com/jobs/search/?geoId=102257491&keywords=teaching&location=London%2C%20England%2C%20United%20Kingdom')
driver.implicitly_wait(5)
```

```python
try:
    x = driver.find_element_by_xpath('//button[@class="cta-modal__dismiss-btn"]')
    x.click()
except:
    print('ok')
```

```python
data_list = []
```

```python
for i in range(1,20):
    temp_list = []
    element = driver.find_element_by_xpath(f'//div[@data-row="{i}"] | //a[@data-row="{i}"]')
    link = driver.find_element_by_xpath(f'(//div[@data-row="{i}"]/a[@class="base-card__full-link"]) | //a[@data-row="{i}"]')
    title = driver.find_element_by_xpath(f'(//div[@data-row="{i}"]//h3[@class="base-search-card__title"]) | //a[@data-row="{i}"]//h3[@class="base-search-card__title"]')
    company = driver.find_element_by_xpath(f'(//div[@data-row="{i}"]//h4[@class="base-search-card__subtitle"]) | //a[@data-row="{i}"]//h4[@class="base-search-card__subtitle"]')
    location = driver.find_element_by_xpath(f'(//div[@data-row="{i}"]//span[@class="job-search-card__location"]) | //a[@data-row="{i}"]//span[@class="job-search-card__location"]')
    time_element = driver.find_element_by_xpath(f'(//div[@data-row="{i}"]//time) | //a[@data-row="{i}"]//time')
    temp_list.append(title.text)
    temp_list.append(company.text)
    temp_list.append(location.text)
    temp_list.append(time_element.get_attribute('datetime'))
    temp_list.append(link.get_attribute('href'))
    data_list.append(temp_list)
```

```python
data_list
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Title','Company','Location','Time','Link'],data=data_list)
```
